```python
consumo_electrico = pd.read_csv('Consumo de energía eléctrica por entidad federativa.csv')
consumo_electrico
```

```python
df_consumo = consumo_electrico.iloc[14:47,:] #quitar filas sin datos
df_consumo.drop(["Unnamed: 1", "Unnamed: 2", "Unnamed: 3"], axis=1, inplace=True)

df_consumo.rename(columns = {"Unnamed: 4" : "2002",
                           "Unnamed: 5" : "2003",
                           "Unnamed: 6" : "2004",
                           "Unnamed: 7" : "2005",
                           "Unnamed: 8" : "2006",
                           "Unnamed: 9" : "2007",
                           "Unnamed: 10" : "2008",
                           "Unnamed: 11" : "2009",
                           "Unnamed: 12" : "2010",
                           "Unnamed: 13" : "2011",
                           "Unnamed: 14" : "2012",
                           "Unnamed: 15" : "2013",
                           "Unnamed: 16" : "2014",
                           "Unnamed: 17" : "2015",
                           "Unnamed: 18" : "2016",
                           "Unnamed: 19" : "2017",
                           "Unnamed: 20" : "2018",
                           "Unnamed: 21" : "2019",
                           "Unnamed: 22" : "2020"}, inplace = True)

df_consumo = df_consumo.set_index("Sistema de Información Energética").unstack().reset_index() # Acomodamos la base en formato largo y reiniciamos el índice
df_consumo.columns = ["Año", "Entidad", "Consumo_energético"]


df_consumo.Consumo_energético = df_consumo.Consumo_energético.replace({",":""}, regex=True)

df_consumo.Año = pd.to_numeric(df_consumo.Año)
df_consumo.Consumo_energético = pd.to_numeric(df_consumo.Consumo_energético)

df_consumo.Entidad = df_consumo.Entidad.replace(["Total Nacional 1", "Distrito Federal", "Estado de México" ],["Nacional", "Ciudad de México", "México"])

df_consumo
```

```python
# Solo consideramos del años 2002 a 2019 ya que los registros para el 2020 del consumo energético están incompletos
#serie_consumo = df_consumo[(df_consumo.Año <= 2019)].groupby("Año")["Consumo_energético"].mean()
serie_consumo = df_consumo.groupby("Año")["Consumo_energético"].mean()

# Gráfica del historico del consumo energético
fig = px.line(serie_consumo, title="Histórico del Consumo Enegético de 2002 a 2019")
fig.update_traces(mode='lines+markers', marker_line_width=3, marker_size=10, marker_color='rgba(10, 550, 19, 90)')
fig.update_xaxes(title_text="Año")
fig.update_yaxes(title_text="<b>Consumo Energético</b>")
fig.show()
```

```python
fig = px.line(df_consumo[df_consumo.Entidad != "Nacional"], x="Año", y="Consumo_energético", color="Entidad", title="Histórico de la intensidad energética nacional (2002-2019)")
fig.update_xaxes(title_text="Año")
fig.update_yaxes(title_text="<b>Intensidad energética</b> ()")
fig.show()
```
